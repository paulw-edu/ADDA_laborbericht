```python
# sample data
import os
import shutil

subdirs = os.listdir("data")
baumsorten = [d for d in subdirs]
for sorte in baumsorten:
  bilder = os.listdir("data/"+sorte)
  for bild in bilder[:3]:
    path = "barkImages/"+sorte+"/"
    if not os.path.exists(path):
      os.makedirs(path)
    shutil.copy("data/"+sorte+"/"+bild, path)

```
